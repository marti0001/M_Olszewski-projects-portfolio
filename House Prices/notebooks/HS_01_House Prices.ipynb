##  Przeanalizowanie zadania
Moim zadaniem jest stworzenie modelu który bedzie przewidywał ceny kazdego domu na podstawie dostarczonego zbioru danych który zawiera róznorodne cechy związane z domami.

Zmienna docelowa ceny domów jest to wartość ciągła wiec zadanie jest problemem regresji

Metryka oceny bedzie pierwiastek błedu średniokwadratowego ( RMSE ) który jest średnią kwadratową róznic miedzy wartościami obserowanymi a wartościami przewidywanymi. Dowiadujemy sie dzeki niemu, w jakim stopniu model myli sie w przewidywaniach, wraz ze wzrostem wartości błedu rośnie waga tego wskaźnika

#### Rozkład zmiennej docelowej

```python
plt.figure(figsize=(10,6))
sns.histplot(train['SalePrice'],kde = True, bins= 30)
plt.title('Rozkład SalePrice')
plt.xlabel('SalePrice')
plt.ylabel('Częstość')
plt.show()
```

Wnioski

- Rozkład jest prawostronnie asymetryczny (prawoskośny), co oznacza ze wieksza  cześć danych koncentruje sie w niższych wartosciach od 100 000 do 250 000
- Rozkład jest jednomodalny poniewaz posiada jeden wyraźny szczyt w okolicach wartości 150 000- 200 000, oznacza to ze większość cen domów w zbiorze danych mieści sie w tym przedziale

Asymetryczność ma negaywny wpływ na działanie niekótrych modeli regresji,takich jak regrsja liniowa które zakładaja normalność rozkładu zmiennej docelowej.

```python
#train['LogSalePrice'] = np.log(train['SalePrice'])
train['LogSalePrice'] = np.log1p(train['SalePrice']) # logarytm naturalny
sns.histplot(train['LogSalePrice'], kde=True, bins=30)
```

#### Wnioski

- Rozład jest bardziej symetryczny i zbliżony do normalnego (rozkładu Gaussa), oznacza to ze dane po transformacji lepiej spełniaja założenia normalnosci co wpłynie korzystniej na algorytmy uczenia maszynowego, takich jak regresja liniowa.
- wysokie ceny, które wczesniej znajdowały sie na prawym koncu rozkładu i wpływaja na asymetrie zostały skompresowane w bardzie zwartej formie przez co ich wpływ na modelowanie jest teraz ograniczony
- dominujacym przedziałem jest 12 co odpowiada rzeczywistej wartosci ceny
- rozkład normalny jest zblizony do normalnego. zastosowanie tej transrmacji poprawi predykcje

Funkcja wykonuje logarytm naturalny (logarytm w podstawie e) dodoając 1 do wartości wejściowej przed wykonaniem logarytmu co pozwala uniknąc błedu logarytmowania wartości 0 (prowadzi do nieskonczoności )

## Pipeline

Pipeline jeststruktura która pozwala automotyzowac i łaczyc rózne etapy przetwarznia danych i trenowania modelu w spójny sposób

Łaczy kolejne kroki przetwarzania danych.
W każdym kroku stosowana jest transformacja a wynik jednego kroku jest przekazywanydo nastepnego 

```python
from sklearn.compose import make_column_transformer, ColumnTransformer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, OneHotEncoder
```

```python
num_train = train.dtypes[train.dtypes != 'object']
num_train.keys()
```

```python
# Wybieram kolumny numeryczne 
num_attribs = ['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual',
       'OverallCond', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'ExterQual',
       'ExterCond', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
       'BsmtFinSF1', 'BsmtFinType2', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF',
       'HeatingQC', 'CentralAir', '1stFlrSF', '2ndFlrSF', 'LowQualFinSF',
       'GrLivArea', 'BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath',
       'BedroomAbvGr', 'KitchenAbvGr', 'KitchenQual', 'TotRmsAbvGrd',
       'Fireplaces', 'FireplaceQu', 'GarageYrBlt', 'GarageFinish',
       'GarageCars', 'GarageArea', 'GarageQual', 'GarageCond', 'PavedDrive',
       'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch',
       'ScreenPorch', 'PoolArea', 'PoolQC', 'Fence', 'MiscVal', 'MoSold',
       'YrSold']
```

```python
object_train = train.dtypes[train.dtypes == 'object']
object_train.keys()
```

```python
# wybieram kolumny kategoryczne
ohe_attribs = ['MSZoning', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'Foundation', 'Heating', 'Electrical',
       'Functional', 'GarageType', 'MiscFeature', 'SaleType', 'SaleCondition']
```

```python
# tworzenie dla potoku dla cech numerycznych
num_pipeline =  Pipeline(steps=[
    ('impute', SimpleImputer(strategy='mean')), # imputacja, wypełnia brakujace wartosci srednia
    ('scaler',StandardScaler()) # standaryzuje dane (skaluje je aby miały srednia 0 i odchylenie standardowe 1)
])
```

```python
ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')), # imputacja, wypełnia brakujeace wartosci najczesciej wystepujaca
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False)) # zwraca wynik w formacie tablicy gestej i ignoruje nieznane wartosci 
])
```

```python
# Łączenie przetworzen dla cech numerycznych i kategorycznych
col_transform = ColumnTransformer(transformers=[
    ('num_p',num_pipeline,num_attribs), # dla kolumn numerycznych 'num_attribs' stosowany jest num_pipeline
    ('ohe_p',ohe_pipeline, ohe_attribs) # dla kolumn kategorycznych 'ohe_attribs' stosowany jest ohe_pipeline
], remainder= 'passthrough') # wszystkie kolumny które nie sa wymienione, są pozostawione bez zmian
```

```python
# tworzenie głównego potoku 
pipeline = Pipeline(steps=[
    ('preprocessing', col_transform) # przetwarza dane za pomoca col_transform
])
```

```python
X = train.drop(['Id','SalePrice', 'LogSalePrice'], axis=1)
y = train['LogSalePrice']
```

```python
X_preprocessed = pipeline.fit_transform(X) 
# fit uczy sie wymaganych parametrów dla przekształcen (np.srednia w SimpleImputer)
# transform przekształca dane zgodnie z wyuczonymi regułami
```

### GridSearchCV

GridSearchCV to narzedzie do automatycznego przeszukiwania wszystkich mozliwych kombinacji siatki hiperparametrów i wyboru najlepszego zestawu na podstawie zdefiniowanej metryki 
- model model dla którego przeszukujemy hiperparametry 
- param_grid siatka hiperparametrów do przeszukiwania
- cv liczba foldów w walidacji krzyżowej
- scoring metryka oceny
- n_jobs wykorzystuje wszystkie dostepne procesy do przyspieszenia obliczeń.

GridSearchCV zawsze maksymalizuje wyniki metryki (scoring)
w przypadku metryki takiej jak dokładnosc które naturalnie rosna ma to sens wieksza wartosc oznacza lepszy wynik 

Jednak dla błędów takich jak MSE, RMSE chcemy minimalizowac wynik (mniejszy błąd oznacza lepszy model ).
Aby umozliwic mnimalizajce błedu przy maksymalizacji wyniku narzedzie GridSearchCV uzywa negatywnej werski metryki MSE( maksymalizacja negatywnego błedu kwadratowego (-MSE) jest róznoważna minimalizajci samego MSE)

RMSE (ROOT MEAN SQUARED ERROR) jest pierwiastkiem kwadratowym MSE, dzieki czemu wraca do tych samych jednostek co zmienna docelowe

### RandomizedSearchCV

RandomizedSearchCV to mechanizm optymalizacji modeli którego celem jest znalezienie najlepszych wartości hiperparametrów poprzez losowe przeszukiwanie okreslonego zakresu wartosci.

Działą podobnie do GridSearchCV ale zamiast sprawdzac wszystki mozliwe kombinacje hiperparametrów  wybiera ich losowy podzbiór co znacznie skraca czas obliczen 

#### **RandomForestRegressor** 

```python
random_forest = RandomForestRegressor(random_state=42)
```

```python
param_grid_rfr = {
    'n_estimators': randint(low = 400, high = 600),           # Liczba drzew
    'max_depth': randint(low = 10, high = 20),             # Maksymalna głębokość drzewa
    'min_samples_split': randint(low = 2, high = 4),             # Minimalna liczba próbek do podziału węzła
    'min_samples_leaf': randint(low = 1, high = 4),               # Minimalna liczba próbek w liściu
    'bootstrap': [True, False]                  # Czy losowe próbkowanie danych (bootstrap)
}
```

```python
random_search_rfr = RandomizedSearchCV(
    estimator = random_forest,
    param_distributions= param_grid_rfr,
    n_iter= 100,
    cv= 5,
    random_state= 42,
    scoring='neg_mean_squared_error',
#    n_jobs= -1,
    verbose= 2
)
```

```python
random_search_rfr.fit(X_train, y_train)
```

```python
np.sqrt(-1 * random_search_rfr.best_score_)
```

#### **XGBRegressor** 

```python
param_grid_xgb = {
    'n_estimators': randint(low =200, high = 400), # oznacza ilosc drzew
    'learning_rate': uniform(0.04, 0.06), # krok uczenia sie(tempo uczenia sie mniejsza wartosc oznacza wolniejsze ale bardziej stabilne uczenie)
    'max_depth': randint(low =2 , high = 4), # maksymalna głebokosc każdego drzewa wartosc None oznacza brak oganiczenia 
    'min_child_weight': uniform(0.5, 1.5), #minimalna waga obserwacji dla danego węzła
    'subsample': uniform(0.6, 0.4), # podzbiór na którym model uczy sie (wtedy jak jest overfitting, zmniejsz z 1.0 do 0.8 czy nawet do 0.6).
    'colsample_bytree': uniform(0.9, 0.7), # podobnie jak sample, tylko na kolumnach (cechach). Zmniejsz wtedy gdy pojawia sie overfitting
    'reg_alpha': uniform(0, 0.1), # regularyzacja L1, pomaga kontrolowac złozonośc modelu
    'reg_lambda': randint(low =1, high = 2) # regularyzacja L2 
}
```

```python
xgb_model_rand = XGBRegressor(random_state=42,verbosity =0)

random_search_xgb = RandomizedSearchCV(
    estimator=xgb_model_rand, #model który bedzie optymalizowany 
    param_distributions=param_grid_xgb, #przestrzen hiperparametrów do przeszukania 
    n_iter=100, #liczba losowy kombinacji parametrów do przetestowania 
    scoring='neg_mean_squared_error', # kryterium oceny modelu
    random_state=42, # ziarno dla powtarzalnych wyników
    n_jobs=-1, # pozwala na uzycie wszystkich dostepnych procesów do równoległego przetwarzania
    cv=5 
)
```

```python
random_search_xgb.fit(X_train, y_train)
```

```python
np.sqrt(-1 * random_search_xgb.best_score_)
```
