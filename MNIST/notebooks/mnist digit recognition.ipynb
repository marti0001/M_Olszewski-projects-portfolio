## Zastosowanie modelu w praktyce 

Zamierzam przetestowac model który wytrenowałem na zbiorze danych MNIST uzywając własnorecznie pisanych liczbach.

Napisałem własnoręcznie liczby które pomogą mi sprawdzic model.


```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import load_model
import cv2
```

#### Wczytywanie modelu 

```python
model = load_model('../models/MNIST_model.keras')
```

### Funkcja przetworzenia obrazu

```python
def prepare_image(image_path):
    # Załaduj obraz
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    # Przeskaluj do rozmiaru 28x28
    img = cv2.resize(img, (28, 28))
    # Normalizacja wartości pikseli
    # img = 255.0 - img #inwersja kolorów w zdjecia z białego na czarny i odwrotnie
    img = img / 255.0
    # Zmiana kształtu na (1, 28, 28,1) - wymagany przez model
    img = img.reshape(1, 28, 28,1)
    return img
```

### Przykład wczytania i przetworzenia obrazu

```python
image_path = '../input/num.jpg'  # przykładowa moja liczba  
prepared_image = prepare_image(image_path)
plt.imshow(prepared_image[0, :, :], cmap='gray')

plt.show()
```

```python
# Dokonanie prognozy
prediction = model.predict(prepared_image)
predicted_class = np.argmax(prediction, axis=1)
print(f'Predicted class: {predicted_class[0]}')

plt.imshow(prepared_image.reshape(28, 28), cmap='gray')
plt.title(f'Predicted Class: {predicted_class[0]}')
plt.show()
```

### Wnioski 

Stworzony model konwolucyjne sieci neuronowej sprawdził sie na własnorecznie pisanych cyfrach.

Przetestowanie modelu było dobry testem praktycznym. Model dobrze poradził sobie z testem praktycznym  którym była klasyfikacja obrazu co swiadczy o jego uniwersalnosci.
